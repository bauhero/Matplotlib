```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data=pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data=pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
merge_data=pd.merge(clinical_trial_data,mouse_drug_data,on="Mouse ID",)

# Display the data table for preview
merge_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
grouped_data=merge_data.groupby(["Drug", "Timepoint"]).mean()
# Convert to DataFrame
grouped_df=pd.DataFrame(grouped_data["Tumor Volume (mm3)"]).reset_index()
# Preview DataFrame
grouped_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
grouped_df_sem=merge_data.groupby(["Drug","Timepoint"]).sem()

# Convert to DataFrame
grouped_df_sem=pd.DataFrame(grouped_df_sem["Tumor Volume (mm3)"]).reset_index()

#Convert sem table to pivot
grouped_df_sem_pivot=grouped_df_sem.pivot(index='Timepoint', columns='Drug',values='Tumor Volume (mm3)')

# Preview DataFrame
grouped_df_sem.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
grouped_pivot=grouped_df.pivot(index='Timepoint',columns='Drug',values='Tumor Volume (mm3)')

# Preview that Reformatting worked
grouped_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
fig, ax = plt.subplots()

x_axis=[x*5 for x in range(10)]
ax.scatter(x_axis, grouped_pivot["Capomulin"], marker="^", facecolors="blue")
ax.scatter(x_axis, grouped_pivot["Infubinol"], marker="o", facecolors="red")
ax.scatter(x_axis, grouped_pivot["Ketapril"], marker="d", facecolors="green")
ax.scatter(x_axis, grouped_pivot["Placebo"], marker="s", facecolors="black")

ax.errorbar(x_axis, grouped_pivot["Capomulin"], yerr=grouped_df_sem_pivot["Capomulin"], color="blue", linestyle="dotted", label=None)
ax.errorbar(x_axis, grouped_pivot["Infubinol"], yerr=grouped_df_sem_pivot["Infubinol"], color="red", linestyle="dotted", label=None)
ax.errorbar(x_axis, grouped_pivot["Ketapril"], yerr=grouped_df_sem_pivot["Ketapril"], color="green", linestyle="dotted", label=None)
ax.errorbar(x_axis, grouped_pivot["Placebo"], yerr=grouped_df_sem_pivot["Placebo"], color="black", linestyle="dotted", label=None)

fig.suptitle("Tumor Response to Treatment")

ax.set_xlim(-2, 48)
ax.set_ylim(30, 78)

ax.set_xlabel("Time (Days)")
ax.set_ylabel("Tumor Volume (mm3)")

ax.grid()
ax.legend(loc="best", fontsize="small", fancybox=True)

# Save the Figure

fig.savefig("Tumor Volume (mm3)")
```

```python
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice=merge_data.groupby(["Drug","Timepoint"]).count()

# Convert to DataFrame
mice_count=pd.DataFrame(mice["Mouse ID"]).reset_index()
mice_count=mice_count.rename(columns={"Mouse ID":"Mouse Count"})

# Preview DataFrame
mice_count.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mice_count_pivot=mice_count.pivot(index='Timepoint', columns='Drug',values="Mouse Count")

# Preview the Data Frame
mice_count_pivot.head()
```

```python
# Store the Standard Error associated with Mice Count Grouped by Drug and Timepoint
grouped_count_sem=merge_data.groupby(["Drug","Timepoint"]).sem()

# Convert to DataFrame
count_sem=pd.DataFrame(grouped_count_sem["Mouse ID"]).reset_index()

# Minor Data Munging to Re-Format the Data Frames
count_sem_pivot=count_sem.pivot(index='Timepoint', columns='Drug',values="Mouse ID")

# Preview the data
# count_sem_pivot.head()
```

```python
# Convert the count to percentage for 4 treatments required
mice_count_pivot["Capomulin"]=mice_count_pivot["Capomulin"]/mice_count_pivot["Capomulin"][0]*100
mice_count_pivot["Infubinol"]=mice_count_pivot["Infubinol"]/mice_count_pivot["Infubinol"][0]*100
mice_count_pivot["Ketapril"]=mice_count_pivot["Ketapril"]/mice_count_pivot["Ketapril"][0]*100
mice_count_pivot["Placebo"]=mice_count_pivot["Placebo"]/mice_count_pivot["Placebo"][0]*100

# Generate the Plot (Accounting for percentages)

fig, cx = plt.subplots()

x_axis=[x*5 for x in range(10)]
cx.scatter(x_axis, mice_count_pivot["Capomulin"], marker="o", facecolors="red")
cx.scatter(x_axis, mice_count_pivot["Infubinol"], marker="^", facecolors="blue")
cx.scatter(x_axis, mice_count_pivot["Ketapril"], marker="s", facecolors="green")
cx.scatter(x_axis, mice_count_pivot["Placebo"], marker="d", facecolors="black")

cx.errorbar(x_axis,mice_count_pivot["Capomulin"] , yerr=count_sem_pivot["Capomulin"], color="red",linestyle="dotted",label=None)
cx.errorbar(x_axis,mice_count_pivot["Infubinol"] , yerr=count_sem_pivot["Infubinol"], color="blue",linestyle="dotted",label=None)
cx.errorbar(x_axis,mice_count_pivot["Ketapril"] , yerr=count_sem_pivot["Ketapril"], color="green",linestyle="dotted",label=None)
cx.errorbar(x_axis,mice_count_pivot["Placebo"] , yerr=count_sem_pivot["Placebo"], color="black",linestyle="dotted",label=None)

fig.suptitle("Survival During Treatment")

cx.set_xlim(-2, 48)
cx.set_ylim(32, 102)

cx.set_xlabel("Time (Days)")
cx.set_ylabel("Survival Rate (%)")

cx.grid()
cx.legend(loc="best", fontsize="small", fancybox=True)

# Save the Figure
fig.savefig("Survival During Treatment")

# Show the Figure
plt.show()
```

```python
# Calculate the percent changes for each drug
drugs=grouped_df["Drug"].unique()
result_table=[]

for x in range(len(drugs)):
    result=(grouped_pivot.iloc[-1,x]-grouped_pivot.iloc[0,x])/grouped_pivot.iloc[0,x]*100
    result_table.append(result)
    
# Display the data to confirm
percent_change=pd.DataFrame({"Drug":drugs,"Percent Changes":result_table})
percent_change=percent_change.set_index("Drug")
percent_change
```

```python
# Store all Relevant Percent Changes into a Tuple
percent_change=percent_change.reset_index()
query=["Capomulin","Infubinol","Ketapril","Placebo"]
tuples=tuple(percent_change.loc[percent_change["Drug"]==q].values for q in query)


# Splice the data between passing and failing drugs
x_labels=[]
x_labels_fail=[]
y_values_pass=[]
y_values_fail=[]
for val in tuples:
    x_labels.append(val[0][0])
    if val[0][1]<0:
        y_values_pass.append(val[0][1])

    else:
        y_values_fail.append(val[0][1])



# Orient widths. Add labels, tick marks, etc. 
width = 0.9
ind = np.arange(4)

fig, ax = plt.subplots()

rects1 = ax.bar([0], y_values_pass, width, color='g')
rects2 = ax.bar([1,2,3], y_values_fail, width, color='r')

ax.set_ylabel('%Tumor Volume Change')
ax.set_title('Tumor Change Over 45 Day Treatment')
ax.set_xticks(ind)
ax.set_xticklabels(x_labels)
ax.grid()

# Use functions to label the percentages of changes
def autolabel(rects, ax):
    
# Get y-axis height to calculate label position from.
    (y_bottom, y_top) = ax.get_ylim()
    y_height = y_top - y_bottom

    for rect in rects:
        height = rect.get_height()

        # Fraction of axis height taken up by this rectangle
        p_height = (height / y_height)
# If we can fit the label above the column, do that;
# otherwise, put it inside the column.
        if p_height > 0.1: # arbitrary; 95% looked good to me.
            label_position = (y_height * 0.05)
        else:
            label_position = - (y_height * 0.1)
        ax.text(rect.get_x() + rect.get_width()/2., label_position,'%.0f' % int(height)+ "%",ha='center', va='bottom')

# Call functions to implement the function calls
autolabel(rects1, ax)
autolabel(rects2, ax)

# Save the Figure
fig.savefig("Tumor Change Over 45 Day Treatment.png")

# Show the Figure
fig.show()
```
